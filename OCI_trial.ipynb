```python
# import necessary libraires
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
iris_data = pd.read_csv("drive/MyDrive/Iris.csv")
iris_data.head()
```

```python
x = iris_data.drop(['Species', 'Id'], axis=1)
y = iris_data['Species']

display(x.head())
display(y.head())
```

```python
# create a ml model
model = LogisticRegression().fit(x.values, y)
# Now train the model
#model
```

```python
# Predict using trained model
pred = model.predict([[6.1, 5.5, 7.4, 0.4]])
print(pred)
```
